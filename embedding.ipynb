```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'we are the learning the nlp'
]
```

```python
#define the the vocabulary size
voc_size = 10000
```

```python
#one hot representation for every word in the sentences
one_hot_resp = [one_hot(words,voc_size) for words in sentences]
one_hot_resp
```

```python
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
max_len = 8

post_padding = pad_sequences(one_hot_resp, maxlen=max_len, padding='post')
pre_padding = pad_sequences(one_hot_resp, maxlen=max_len, padding="pre")
```

```python
#feature resprentation of vector
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length = max_len))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(post_padding)
```
